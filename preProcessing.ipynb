```python
data=data.dropna()
```

```python
import pandas as pd

dummies = pd.get_dummies(data['ever_married'], prefix='category')

data = pd.concat([data, dummies], axis=1)

data.drop('ever_married', axis=1, inplace=True)

data.head()

```

```python
import pandas as pd

dummies = pd.get_dummies(data['gender'], prefix='category')

data = pd.concat([data, dummies], axis=1)

data.drop('gender', axis=1, inplace=True)

data.head()
```

```python
import pandas as pd

dummies = pd.get_dummies(data['work_type'], prefix='category')

data = pd.concat([data, dummies], axis=1)

data.drop('work_type', axis=1, inplace=True)

data.head()
```

```python
import pandas as pd

dummies = pd.get_dummies(data['Residence_type'], prefix='category')

data = pd.concat([data, dummies], axis=1)

data.drop('Residence_type', axis=1, inplace=True)

data.head()
```

```python
import pandas as pd

dummies = pd.get_dummies(data['smoking_status'], prefix='category')

data = pd.concat([data, dummies], axis=1)

data.drop('smoking_status', axis=1, inplace=True)

data.head()
```

```python
data.to_csv('Clean_Data_stroke', index=False)
```
